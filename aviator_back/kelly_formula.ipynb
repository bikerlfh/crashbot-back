```python
def kelly_capital(b, p, capital):
    f = (b * p - (1 - p)) / b
    f_capital = capital * f
    return f_capital

def kelly_adapte(b, p, R, capital):
    f = (b * p - (1 - p)) / b
    f2 = f * (1 + R);
    f_capital = capital * f2
    return f_capital
```

```python
kelly_capital(1.8, 0.6, 50000)
```

```python
kelly_adapte(1.8, 0.6, 0.5, 50000)
```
