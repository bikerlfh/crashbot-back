```python
from cryptography.fernet import Fernet


# Generar una clave aleatoria para AES
clave_aes = Fernet.generate_key()

# Inicializar el objeto Fernet con la clave generada
fernet = Fernet(clave_aes)

# Texto a encriptar
texto_original = "Mi clave secreta es 1234"

# Encriptar el texto
texto_encriptado = fernet.encrypt(texto_original.encode())

# Mostrar el texto encriptado
print("Texto encriptado:", texto_encriptado)

# Desencriptar el texto
texto_desencriptado = fernet.decrypt(texto_encriptado).decode()

# Mostrar el texto desencriptado
print("Texto desencriptado:", texto_desencriptado)
```
