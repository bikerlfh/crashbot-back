```python
def find_matches(a, b):
    matches = []
    i = 0
    j = 0
    while i < len(a) and j < len(b):
        if a[i] == b[j]:
            matches.append(a[i])
            i += 1
            j += 1
            continue
        i += 1
        len_matches = len(matches)
        if 3 <= len_matches == a.index(matches[-1]):
            break
        matches = []
    if len(matches) > 0:
        return b[j:]
    return b
```

```python
a = [2,3,4,5,6,7,3,1,5,6,56,4,223,13,45]
b = [7,3,1,5,6,56,4,223,13,45,6,75,4,3]
```

```python
find_matches(a, b)
```
