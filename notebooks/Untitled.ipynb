```python
def siguiente_numero(lista_inicial, historicos):
    # Crear un diccionario para almacenar el último índice de cada número
    indices = {num: -1 for num in lista_inicial}
    print(f"indices: {indices}")
    # Iterar sobre la lista de históricos
    for i in range(len(historicos)):
        # Si el número está en la lista inicial, actualiza el último índice
        if historicos[i] in indices:
            indices[historicos[i]] = i
    print(f"historicos: {historicos}")
    print(f"indices: {indices}")
    # Encontrar el número en la lista inicial con el valor más pequeño (es decir, el que no ha salido en más tiempo)
    numero_siguiente = min(indices, key=indices.get)

    return numero_siguiente
```

```python

# Ejemplo de uso
lista_inicial = [1, 2, 3, 4, 5,50]
historicos = [1, 2, 3, 2, 1, 3, 4, 2, 5, 1, 2, 4, 3, 5, 4, 2,1]

print(siguiente_numero(lista_inicial, historicos))  # Debería imprimir 5

```
